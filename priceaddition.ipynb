```python
import pandas
import requests
pandas.set_option('display.max_columns', None)
```

```python
price_data = pandas.read_csv('../data/combined_nodal_prices.csv')
```

```python
#promote headers 1,2 to be column names by concatenating them
thing = price_data.iloc[0] + '_' + price_data.iloc[1]
print(thing)
price_data.columns = thing
price_data = price_data.iloc[2:]
price_data.head()
```

```python
#concatenate the data from Zone_Data and every column after WestGreenfield together
# the goal is to have a dataframe with datetime and all the zone prices. the year column is all over the place


column_names = price_data.columns.tolist()


column_names[1] = 'Hour_of_Day'

price_data.columns = column_names



# make sure is string
price_data['Zones_Date'] = price_data['Zones_Date'].fillna("").astype(str)

# loop through all bad columns nan.
for x in range(17, len(price_data.columns)):
    col_str = price_data.iloc[:, x].fillna("").astype(str)  # handle NaNs safely
    price_data['Zones_Date'] = price_data['Zones_Date'] + " " + col_str

# clean up extra spaces
price_data['Zones_Date'] = price_data['Zones_Date'].str.strip()

# drop old broken columns


```

```python
import numpy as np
price_data = price_data.drop(columns=np.nan)
price_data['Zones_Date']
```

```python
# display all years
price_data['Zones_Date'] = pandas.to_datetime(price_data['Zones_Date'], errors= 'coerce')
price_data['Zones_Date'].dt.year.unique()
# drop rows where date is nan
price_data = price_data.dropna(subset=['Zones_Date'])
```

```python
price_data.head()
```

```python
price_data.to_csv("prices_zonal.csv")
```
